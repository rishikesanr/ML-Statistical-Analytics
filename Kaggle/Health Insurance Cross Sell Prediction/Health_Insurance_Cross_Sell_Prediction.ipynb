```python
# Distributions of categorical colummns 
cat_cols = ['Gender', 'Driving_License','Previously_Insured','Vehicle_Age','Vehicle_Damage']

for col in cat_cols:
    df[col].value_counts().plot(kind='bar',title=f"Distribution of {col}")
    plt.show()
```

```python
# Distributions of numeric colummns 
num_cols = ['Age', 'Region_Code','Annual_Premium','Policy_Sales_Channel','Vintage']

for col in num_cols:
    df[col].plot(kind='hist',bins=10,title=f"Distribution of {col}")
    plt.show()
```

```python
# Distribution of Response Variable 

df['Response'].value_counts().plot(kind='bar',title='Distribution of Response')

#There is a class imbalance
```
